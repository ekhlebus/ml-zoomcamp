Pandas is a Python library for data manipulation and analysis.

It provides powerful tools to work with two main data structures:

* **DataFrames** – two-dimensional labeled data (table or spreadsheet)
* **Series** – one-dimensional labeled data (every column of a DataFrame)

````python
import numpy as np
import pandas as pd
````

## DataFrames

````python
data = [
    ['Nissan', 'Stanza', 1991, 138, 4, 'MANUAL', 'sedan', 2000],
    ['Hyundai', 'Sonata', 2017, None, 4, 'AUTOMATIC', 'Sedan', 27150],
    ['Lotus', 'Elise', 2010, 218, 4, 'MANUAL', 'convertible', 54990],
    ['GMC', 'Acadia',  2017, 194, 4, 'AUTOMATIC', '4dr SUV', 34450],
    ['Nissan', 'Frontier', 2017, 261, 6, 'MANUAL', 'Pickup', 32340],
]

columns = [
    'Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders',
    'Transmission Type', 'Vehicle_Style', 'MSRP'
]
````

````python
df = pd.DataFrame(data, columns=columns)
````

## Index


**Indexes in Pandas** are labels that uniquely identify rows (and sometimes columns) in a Series or DataFrame.

They help you access, align, and organize data efficiently — similar to row numbers in a spreadsheet, but much more powerful since they can be custom labels (like dates, names, or IDs).

````python
df.index
````

````python
# All series in dataframes have same indexes
df.Make.index
````

````python
# Using indexes we can access rows in dataframe
df.loc[2]  # third row
````

````python
df.loc[[1, 2]]  # second and third row
````

````python
# We can set custom indexes (replace default integer indexes)
df.index = ['a', 'b', 'c', 'd', 'e']
````

````python
df
````

````python
# Access rows using positional indexing
# Here instead of loc we use iloc
df.iloc[[1, 2, 4]]
````

````python
# Reset custom indexes to default integer indexes
df = df.reset_index(drop=True)
````

````python
df
````

````python
df['Engine HP'] * 2
````

## Filtering

````python
df[df['Year'] >= 2010]
````

````python
df[
    df['Make'] == 'Nissan'
]
````

````python
df[
    (df['Make'] == 'Nissan') & (df['Year'] >= 2015)
]
````

## String operations

````python
# Get all 'Vehicle_Style' values. It would be great to standardize the capitalization
df['Vehicle_Style']
````

````python
df['Vehicle_Style'].str.lower()
````

````python
# Replace spaces with underscores
'machine learning zoomcamp'.replace(' ', '_')
````

````python
# Standardize 'Vehicle_Style' values by replacing spaces with underscores and converting to lowercase
df['Vehicle_Style'] = df['Vehicle_Style'].str.replace(' ', '_').str.lower()
````

````python
df
````

## Summarizing operations

````python
# Get summary statistics for 'MSRP' column
df.MSRP.describe()
````

````python
# Get summary statistics for all NUMERIC columns, rounded to 2 decimal places
df.describe().round(2)
````

````python
# Get number of unique values in 'Make' column
df.Make.nunique()
````

````python
# Get number of unique values in all columns, applied for the whole dataframe
df.nunique()
````

## Missing values


````python
# In ML we often need to know how many missing values are there in each column and do something about them
# Calculate column-wise number of missing values (NaNs in each column)
df.isnull().sum()
````

## Grouping


```
SELECT 
    transmission_type,
    AVG(MSRP)
FROM
    cars
GROUP BY
    transmission_type
```

````python
df.groupby('Transmission Type').MSRP.max()
````

````python
df.groupby('Transmission Type').MSRP.describe()
````
